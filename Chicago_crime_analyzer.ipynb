```python
df.drop(['Ward', 'Community Area'], axis=1, inplace = True)
```

```python
df['X Coordinate'].fillna(df['X Coordinate'].mode().values[0], inplace = True)
df['Y Coordinate'].fillna(df['Y Coordinate'].mode().values[0], inplace = True)
```

```python
df['Latitude'].fillna(df['Latitude'].mode().values[0], inplace = True)
df['Longitude'].fillna(df['Longitude'].mode().values[0], inplace = True)
df['Location Description'].fillna(df['Location Description'].mode().values[0], inplace = True)
```

```python
df['Location'].fillna(df['Location'].mode().values[0], inplace = True)
```

```python
df['month'] = df['Date'].dt.month
df['weekday'] = df['Date'].dt.dayofweek

```

```python
df['hour'] = df['Date'].dt.hour
```

```python
df['quarter'] = df['Date'].dt.quarter
```

```python
df.isnull().sum()
```

```python
# Categorizing crimes as severe or non-severe 

severe_crimes = ['HOMICIDE', 'ASSAULT', 'BATTERY', 'ROBBERY', 'CRIM SEXUAL ASSAULT']
df['Severity'] = df['Primary Type'].apply(lambda x: 'Severe' if x in severe_crimes else 'Non Severe')
```

```python
# Define the seasons based on the month
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'

# Apply the function to create a 'Season' column
df['Season'] = df['month'].apply(get_season)
```
